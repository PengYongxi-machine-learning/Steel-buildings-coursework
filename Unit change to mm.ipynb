```python
import pandas as pd
df = pd.read_excel('Section.xlsx')
print(df.columns)
df['Iy (mm^4)'] = df['Iy (cm^4)']*10**4
df['Iz (mm^4)'] = df['Iz(cm^4)']*10**4
df['R_y (mm)'] = df['R_y(cm)']*10
df['R_z (mm)'] = df['R_z(cm)']*10
df['W_el,y (mm^3)'] = df['W_el,y (cm^3)']*10**3
df['W_el,z (mm^3)'] = df['W_el,z (cm^3)']*10**3
df['W_pl,y (mm3)'] = df['W_pl,y (cm3)']*10**3
df['W,pl,z (mm^3)'] = df['W,pl,z (cm^3)']*10**3
df['I_w (mm^6)'] = df['I_w (dm^6)']*100**6
df['I_T (mm^4)'] = df['I_T (cm^4)']*10**4
df['A (mm^2)'] = df['A (cm^2)']*10**2
print(df.columns)
```

```python
df = df.drop(columns = ['Iy (cm^4)','Iz(cm^4)','R_y(cm)','R_z(cm)', 'W_el,y (cm^3)', 'W_el,z (cm^3)', 'W_pl,y (cm3)',
                   'W,pl,z (cm^3)', 'I_w (dm^6)', 'I_T (cm^4)', 'A (cm^2)'])
df.to_csv("section_mm.csv", index=False)
```
